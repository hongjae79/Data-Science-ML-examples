```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import re
```

```python
X = []
Y = []

# some numbers show up as 1,170,000,000 (commas)
# some numbers have references in square brackets after them
non_decimal = re.compile(r'[^\d]+')
```

```python
for line in open('moore.csv'):
    r = line.split('\t') #split the line by tab

    x = int(non_decimal.sub('', r[2].split('[')[0]))
    y = int(non_decimal.sub('', r[1].split('[')[0]))
    X.append(x)
    Y.append(y)
```

```python
#change to numpy array
X = np.array(X)
Y = np.array(Y)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
Y = np.log(Y)
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
#apply the equation for linear regression.  
denominator = X.dot(X)- X.mean() * X.sum() 
a = (X.dot(Y) - X.mean() * Y.sum()) / denominator
b = (Y.mean() * X.dot(X) - X.mean() * X.dot(Y)) /denominator
```

```python
#predicted y
Y_hat = a * X + b
```

```python
#plot them all 
plt.scatter(X,Y)
plt.plot(X, Y_hat, color = 'red')
plt.show()
```

```python
#calculate r-squared
d1 = Y - Y_hat
d2 = Y - Y.mean()
r2 = 1 - d1.dot(d1)/d2.dot(d2)
print("the r-squared is ", r2)
```
