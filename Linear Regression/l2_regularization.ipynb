```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate the data
N = 50 

X = np.linspace(0,10,N)

#numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)[source]
#Return evenly spaced numbers over a specified interval.
#Returns num evenly spaced samples, calculated over the interval [start, stop].
#The endpoint of the interval can optionally be excluded.

Y = 0.5*X + np.random.randn(N)
# randn(d0, d1, …, dn)	Return a sample (or samples) from the “standard normal” distribution.
```

```python
#create outliers
Y[-1] += 30 # add 30 to last one in Y
Y[-2] += 30 # add 30 to second last one in Y
```

```python
# plot the data
plt.scatter(X, Y)
plt.show()
```

```python
# add bias term
X = np.vstack([np.ones(N), X]).T
#numpy.vstack(tup)[source]
#Stack arrays in sequence vertically (row wise).
```

```python
#plot the maximum likelihood solution 
w_ml = np.linalg.solve(X.T.dot(X), X.T.dot(Y))  
Y_hat = X.dot(w_ml)
plt.scatter(X[:,1], Y)
plt.plot(X[:,1], Y_hat)
plt.show()
```

```python
# plot the regularized solution
# probably don't need an L2 regularization this high in many problems
# everything in this example is exaggerated for visualization purposes

l2 = 1000 #lambda
w_map = np.linalg.solve(l2*np.eye(2) + X.T.dot(X), X.T.dot(Y))
Yhat_map = X.dot(w_map)
plt.scatter(X[:,1], Y)
plt.plot(X[:,1], Yhat_map)
plt.show()

#this can avoid outliers by giving penalty on |w|^2
```
